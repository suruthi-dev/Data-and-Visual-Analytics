```python
country = df['Country'].value_counts()
country
```

```python
python_dev = df.groupby("Country")['LanguageWorkedWith'].apply(lambda x : x.str.contains("Python").sum())
python_dev
```

```python
python_users = pd.concat([country,python_dev],axis=1)
python_users
```

```python
python_users.columns = ['respondants','users']
```

```python
python_users
```

```python
python_users['percentage_of_python_users'] = (python_users['users']/python_users['respondants'])*100
python_users
```

```python
python_users.sort_values(by='percentage_of_python_users',ascending=False).head(50)
```
